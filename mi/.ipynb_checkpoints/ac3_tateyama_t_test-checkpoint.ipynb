## 館山さんのデータ「シリンダー、ディスク」を同一データとして扱ってよいか問題を検定で確かめたい。

```python
# ライブラリ
import matplotlib.pyplot as plt
import pandas as pd
import math
import scipy.stats as st
import numpy as np
```

```python
# pdに含まれるread_csvというメゾッドを使ってデータを読みこみdfとして保存
df = pd.read_csv("hr_cylinder_disk_t.csv") 

# dfの中身を確認
print(df)
```

```python
#平均値、標準偏差に大きな違いがない事を確認
#基本統計量を確認する事で、処理すべき妙な値のデータがないか把握する
df.describe()
```

```python
#元データのヒストグラム。外れ値が気になる。
bins = np.linspace(0, 100, 500)
plt.hist(df.Cylinder, bins, alpha = 0.5, label='Cylinder')
plt.hist(df.Disk, bins, alpha = 0.5, label='Disk')
plt.legend(loc='upper left')

plt.show()
```

```python
#分位2％（左側から2％の地点まで）を削除する
#等分散とまではいかないが、分散も近い値になった。
q = df.Cylinder.quantile(0.02)
p = df.Disk.quantile(0.02)

new_df = df.query('Cylinder > @q')
new_df_cylinder = new_df.drop(columns='Disk')

new_df = df.query('Disk > @p')
new_df_disk = new_df.drop(columns='Cylinder')

new_df = new_df_cylinder.append(new_df_disk)
new_df.describe()
```

```python
#外れ値処理後のヒストグラム。
bins = np.linspace(0, 100, 500)
plt.hist(new_df['Cylinder'], bins, alpha = 0.5, label='Cylinder')
plt.hist(new_df['Disk'], bins, alpha = 0.5, label='Disk')
plt.legend(loc='upper left')

plt.show()
```

## 正規分布ではないので、ｔ検定は出来ないが、一応出力。
* t値 ： 平均値の差の大きさの度合い。
* p値 ： 母集団の平均が同じだと仮定したときに、平均の差が、検定を行ったデータより大きくなる確率。この値でもって1%や5%の判定をする。

```python
#p値は低いが、なかなか微妙なラインだと言える。
#有意水準を5%とするとp値は<0.05なので，
#帰無仮説「CylinderとDiskの平均値に差はない」は棄却され、対立仮説「生存者の平均年齢<非生存者の平均年齢」を成立させることになる。
#平均値は酷似していたし、分散も大差ないのに、この結果は納得できない。やるなら正規分布にする必要がある。
t,p_val = st.ttest_ind(new_df_cylinder,new_df_disk,equal_var=True)
print('T=',t)
print('p=',p_val)
```

```python
#BOXCOX変換を行って、分布を正規分布に近づける
from scipy.stats import boxcox
from scipy.special import inv_boxcox
import itertools
```

```python
#cylinderとdiskをそれぞれ別のデータフレームとして用意する。BoxCox変換したいため。
cylinder_cut = pd.DataFrame({'Cylinder':new_df['Cylinder']})
disk_cut = pd.DataFrame({'Disk':new_df['Disk']})

cylinder_cut = cylinder_cut.dropna(how='any')
disk_cut = disk_cut.dropna(how='any')
```

```python
#boxcox変換の最適なパラメータを調査。
#指定なしの場合、自動で最も正規分布に近づけるパラメータが選択される

#Cylinderについて、6種類のlambdaでBoxcox変換
#lambda=16.45が最適だか、Diskも同じ数値で正規分布に近い形を選ばなければならないから、lambda=17を選択

cylinder_trans = cylinder_cut



fig = plt.figure(figsize=(6.0, 6.0))

list_lambda = [0, 1, 5, 10, 16, 17, 20]
for i, i_lambda in enumerate(list_lambda):
    cylinder_trans[ 'val_'+str(i) ] = boxcox(cylinder_trans.Cylinder, lmbda = i_lambda )
    fig.add_subplot(4, 2, i+1).hist(cylinder_trans['val_'+str(i)], bins=20, color='r', alpha=0.5, density=True)
    plt.title("lambda="+str(list_lambda[i]))
    plt.xlabel('val')
    plt.ylabel('density')

cylinder_trans[ 'Cylinder_auto' ], best_lambda = boxcox( cylinder_trans.Cylinder )
fig.add_subplot(4, 2, 8).hist(cylinder_trans['Cylinder_auto'], bins=20, color='r', alpha=0.5, density=True)
plt.title("lambda="+str(round(best_lambda, 2)))
plt.xlabel('Cylinder')
plt.ylabel('density')

fig.tight_layout()
fig.show()
plt.show()
```

```python
#データフレーム確認
cylinder_trans
```

```python
#Diskについて、6種類のlambdaでBoxcox変換 
#lambda=19.1が最適だが、Cylinderも同じ数値で正規分布に近い形を選ばなければならないから、lambda=17を選択

disk_trans = disk_cut



fig = plt.figure(figsize=(6.0, 6.0))

list_lambda = [0, 1, 5, 10, 16, 17, 20]
for i, i_lambda in enumerate(list_lambda):
    disk_trans[ 'val_'+str(i) ] = boxcox(disk_trans.Disk, lmbda = i_lambda )
    fig.add_subplot(4, 2, i+1).hist(disk_trans['val_'+str(i)], bins=20, color='r', alpha=0.5, density=True)
    plt.title("lambda="+str(list_lambda[i]))
    plt.xlabel('val')
    plt.ylabel('density')

disk_trans[ 'Disk_auto' ], best_lambda = boxcox( disk_trans.Disk)
fig.add_subplot(4, 2, 8).hist(disk_trans['Disk_auto'], bins=20, color='r', alpha=0.5, density=True)
plt.title("lambda="+str(round(best_lambda, 2)))
plt.xlabel('Disk')
plt.ylabel('density')

fig.tight_layout()
fig.show()
plt.show()
```

```python
#データフレーム確認
disk_trans
```

```python
#必要なカラムのみ残す
cylinder_trans = cylinder_trans.val_6
disk_trans = disk_trans.val_6
```

```python
#データフレーム確認
cylinder_trans
```

```python
#データセットの値を小数第3位までにしたい
import decimal

decimal.getcontext().prec = 3

tmp_1 = []
tmp_2 = []
n = 2# 切り捨てしたい桁

for i in cylinder_trans:
    m = decimal.Decimal(i)
    num = math.floor(m * 10 ** n) / (10 ** n)
    print(i,"=>",m,"=>",num)
    tmp_1.append(num)
    
for i in disk_trans:
    m = decimal.Decimal(i)
    num = math.floor(m * 10 ** n) / (10 ** n)
    print(i,"=>",num)
    tmp_2.append(num)
    
cylinder_trans_round = pd.DataFrame({'Cylinder':tmp_1})
disk_trans_round = pd.DataFrame({'Disk':tmp_2})
```

```python
#データセットの中身を確
disk_trans_round.describe()
```

```python
#ヒストグラムとして表示してみる
#lambda=17
#bins = np.linspace(0,600000000000000000000000000000000,50)
#lambda=19
bins = np.linspace(0,600000000000000000000000000000000000000,50)
plt.hist(cylinder_trans_round.Cylinder, bins, alpha = 0.5, label='Cylinder')
plt.hist(disk_trans_round.Disk, bins, alpha = 0.5, label='Disk')
plt.legend(loc='upper left')

plt.show()
```

```python
disk_trans_round.describe()
```

```python
cylinder_trans_round.describe()
```

```python
#t検定を行ってみる
st.ttest_ind(cylinder_trans_round.Cylinder,disk_trans_round.Disk,equal_var=False)
```

#### 
BoxCox変換する時に、綺麗な正規分布に近づける為のハイパーパラメータlambdaがありますが、この数値を2群で異なる値にしていたらダメ。
このパラメータ値によって変換後の値のスケールも大きく変わる為、同じパラメータ値にしなければならない。
しかし私は２群それぞれが最も綺麗な正規分布になるパラメータとして、別々のパラメータを振っていました。それが良くなかったです。

lambda=17で統一する事で、何とかスケール合わせが出来た。
これがベストプラクティスではないが、ひとまずここまでできた。

## ノンパラメトリック検定の中でもマン・ホイットニーのU検定(ウィルコクソンの順位和検定と同じ)を行う

```python
from scipy.stats import mannwhitneyu
```

```python
#結果、p < 0.05なので、帰無仮説は棄却され2群の代表値には差があるといえる。
u,pval = mannwhitneyu(new_df_cylinder,new_df_disk,alternative='two-sided')

print('U=',u)
print('p=',pval)
```
