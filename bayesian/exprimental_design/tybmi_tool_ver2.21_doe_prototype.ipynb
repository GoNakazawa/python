#### <font color="white"> 現状のデータセットのD最適基準を確認<font>

```python
def D_optimal_value(df_reg):
    x_generated = df_reg.copy()
    autoscaled_x_generated = ((x_generated - x_generated.min()) / (x_generated.max() - x_generated.min())) * 2 - 1
    xt_x = np.dot(autoscaled_x_generated.T, autoscaled_x_generated)
    number_of_freedom = len(autoscaled_x_generated) - len(autoscaled_x_generated.columns) - 1
    
    if number_of_freedom != 0:
        d_optimal_value = (np.linalg.det(xt_x) ** (1 / number_of_freedom))  / len(autoscaled_x_generated)
    else:
        d_optimal_value = np.linalg.det(xt_x)
    return d_optimal_value
```

```python
# 現状のD最適基準の値を確認
df_exp = df_reg.drop(objectives, axis=1)
d_optimal_value_reg = D_optimal_value(df_exp)
print('D optimal value for a given dataset (0 <= D optimal value <= 1):', d_optimal_value_reg)
```

## 2.2 与えられた範囲で実験を計画

```python
number_of_selecting_samples = 1  # 1度に選択するサンプル数
number_of_random_searches = 250 # ランダムにサンプルを選択して D 最適基準を計算する繰り返し回数
x_generated = samples.copy()
autoscaled_x_generated = ((x_generated - x_generated.min()) / (x_generated.max() - x_generated.min())) * 2 - 1

# 実験条件の候補のインデックスの作成
np.random.seed(22) # 乱数を生成するためのシードを固定
all_indexes = list(range(x_generated.shape[0]))
new_selected_indexes = np.random.choice(all_indexes, number_of_selecting_samples, replace=False)
selected_samples = autoscaled_x_generated.iloc[new_selected_indexes, :]

# D 最適基準に基づくサンプル選択
for number_of_selection in range(int(numer_of_total_experiments / number_of_selecting_samples)):
    for random_search_number in range(number_of_random_searches):
        # 1. ランダムに候補を選択
        new_selected_indexes = np.random.choice(all_indexes, number_of_selecting_samples, replace=False)
        new_selected_samples = autoscaled_x_generated.iloc[new_selected_indexes, :]
        new_selected_samples = pd.concat([selected_samples, new_selected_samples])

        # 2. D 最適基準を計算
        d_optimal_value = D_optimal_value(new_selected_samples)
        
        # 3. D 最適基準が前回までの最大値を上回ったら、選択された候補を更新
        if random_search_number == 0:
            best_d_optimal_value = d_optimal_value.copy()
            selected_sample_indexes = list(new_selected_samples.index).copy()

        else:
            if best_d_optimal_value < d_optimal_value:
                best_d_optimal_value = d_optimal_value.copy()
                selected_sample_indexes = list(new_selected_samples.index).copy()

    selected_samples = autoscaled_x_generated.iloc[selected_sample_indexes, :]
    print('number_of_samples: ', len(selected_samples))

selected_sample_indexes = list(selected_samples.index) # リスト型に変換

# 選択されたサンプル、選択されなかったサンプル
selected_samples = x_generated.iloc[selected_sample_indexes, :].reset_index().drop('index', axis=1)  # 選択されたサンプル
remaining_indexes = np.delete(all_indexes, selected_sample_indexes)  # 選択されなかったサンプルのインデックス
remaining_samples = x_generated.iloc[remaining_indexes, :]  # 選択されなかったサンプル

# 保存
selected_samples.to_csv('selected_samples.csv')  # 選択されたサンプルを csv ファイルに保存。同じ名前のファイルがあるときは上書きされるため注意
remaining_samples.to_csv('remaining_samples.csv')  # 選択されなかったサンプルを csv ファイルに保存。同じ名前のファイルがあるときは上書きされるため注意

print('D optimal value for a new dataset:', best_d_optimal_value) # 相関行列の確認
```

```python
import random
from sklearn import preprocessing
from scipy.spatial import distance

df_exp_copy = df_exp.copy().reset_index()
df_exp_copy['minimum_distance'] = 0
df_exp_copy['minimum_index'] = 0
index_list = list(df_exp_copy.index)
ss = preprocessing.StandardScaler()
ss.fit(df_exp)
df_exp_autoscaled = ss.transform(df_exp)
    
for exp_point_index in index_list:
    distance_list = []
    selected_samples_autoscaled = ss.transform(selected_samples)
    exp_point =df_exp_autoscaled[exp_point_index,]
        
    for sample_index,selected_sample in enumerate(selected_samples_autoscaled):
        dist = distance.euclidean(exp_point, selected_sample)
        distance_list.append(dist)
        
    df_exp_copy.loc[exp_point_index,'minimum_index'] = distance_list.index(min(distance_list))
    df_exp_copy.loc[exp_point_index,'minimum_distance'] = min(distance_list)

#plt.hist(df_exp_copy['minimum_distance'])
```
